```python
 #import sequential api (functional apis??)
from keras.models import Sequential
from keras.layers import *
#Input, Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python


model.add(Conv2D(64,(4,8), activation='relu', padding="same",input_shape=(1,128,64)))
model.add(BatchNormalization())

model.add(UpSampling2D(size=(1,2)))
model.add(Conv2D(64,(8,16), padding="same", activation='relu'))
model.add(BatchNormalization())

model.add(UpSampling2D(size=(1,2)))
model.add(Conv2D(64,(16,32),  padding="same",  activation='relu'))
model.add(BatchNormalization())

model.add(UpSampling2D(size=(1,2)))
model.add(Conv2D(64,(32,64), padding="same",   activation='relu'))
model.add(BatchNormalization())

model.add(UpSampling2D(size=(1,2)))
model.add(DepthwiseConv2D(1, (64,1), activation='relu'))
model.add(BatchNormalization())


```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```
